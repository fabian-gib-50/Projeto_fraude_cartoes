### Criei uma função para poder descrever melhor as variáveis **"def sumarydata"**

   - Checagem de dados `ausentes`, dados nulos ( nan ), dados NaN (Not a Number) representam dados missing, que são registros que não existem ou que se perderam dentro do conjunto da base de dados
   
   - Check das variáveis, types (tipos) distribuição dos valores, e nome das variáveis

```python
# Função para otimizar a leitura das entredas e chegar as variáveis

print( 'Python imprima descritivo das variáveis\n' )

def sumarydata ( data ): 
   
    summary = pd.DataFrame( data.dtypes, columns=[ 'dtypes' ] )
    summary = summary.reset_index()
    summary[ 'Nome Variável' ] = summary['index']
    summary = summary[[ 'Nome Variável', 'dtypes' ]]
    summary[ 'Dados Únicos' ] = data.nunique().values
    summary[ 'Valores Primários' ]  = data.loc[0].values
    summary[ 'Valores Secudários' ] = data.loc[1].values
    summary[ 'Dados Nulos' ] = data.isnull().sum().values
    summary[ 'Dados Nulos (%)' ] = ((( data.isnull().sum() / data.shape[0] ) * 100 ).sort_values(ascending = False ))
    
    return summary

check = sumarydata ( data )
check.head(5)
```

### Aplica do método matemático PCA para Python

  - Aplicação de uma técnica para reduzir a dimensão do DataFrame, ( base de dados ), preservando suas propriedades e integridade dos dados.
   - Vejamos os recursos gerados por este método, como cada recurso se correlaciona com a classe desse DataFrame.

  - Origem e Aplicação: PCA, é um modelo de procedimento `matemático` que utiliza a transformação `ortogonal`, (ortogonalização de vetores) para converter um conjunto de observações de variáveis possivelmente correlacionadas, em conjunto de valores de variáveis `linearmente` não correlacionadas chamadas de componentes principais ).
  
  O método PCA nos permite a redução da dimensionalidade enquanto mantém o maior número possível de informações. Para conseguir isso, o algoritmo encontra um conjunto novo de recursos - os chamados **`componentes`**.

Esses componentes são em número menor or igual às variáveis originais. No caso deste projeto, os componentes achados pela transformação da PCA são as próprias colunas **['V28', 'V27', 'V26', 'V25', 'V24', 'V23', 'V22', 'V20', 'V15', 'V13', 'V8']**.

### Padronização das colunas **Time e Amount**, escala recursal

  - Aplicação do método de padronização por meio da classe `StandardScaler`.
   - Assim que feito o reshape, padronização das variáveis em questão, aplico um clean e split, limpeza e separação, para construção do dados de `treino testes`

```python
# Aplicação da Classe StandardScaler para padronizar as variáveis, "Time-tempo e Amount-quantia", copia dos dados e separação.
# Passando para variável objeto python "dfp"
# Checagem das "features" tratadas (padronizadas) com método "data_clean.iloc[  :3, -7: ]"

dfp = data.copy()

std_scaler = StandardScaler()
dfp[ 'std_amount' ] = std_scaler.fit_transform( dfp[ 'Amount' ].values.reshape(-1, 1))
dfp[ 'std_time' ]   = std_scaler.fit_transform( dfp[ 'Time' ].values.reshape(-1, 1))

dfp.drop([ 'Time', 'Amount' ], axis=1, inplace=True)

dfp.iloc[  :3, -7: ]
```

### PipeLine de Preparação da Base para os trabalhos de criação do modelo de predição com `Machine Learning`

### Processo de separação dos dados para treinamento
  - Em train_X, val_X, train_y, val_y
      - Obs:. Aplicarei um modelo de regressão, porque minha variável de resposta é numérica e meu problema de negócio requer **Regressão Logística**.

```python
# separar variáveis entre X e y
X = dfp.drop( 'Class', axis=1 )
y = data [ 'Class' ]

# dividir o dataset entre treino e teste
train_X, val_X, train_y, val_y = train_test_split(X, y, stratify=y, shuffle=True)
```

### **Balanceamento dos dados**

  - Para balancear os dados, usarei a biblioteca `imblearn`, técnicas de balanceamento de dados para que nosso modelo preditivo esteja o mais próximo da relalidade possível. 
   - Neste projeto usarei *Under-sampling*, procurando preservar as classes minotirárias, classes que compõem o conjunto de variáveis de classe

   - Agora fica mais evitente que não há discrepâncias, como nos gráficos acima, nas explorações iniciais do Dataset.
   - Print das quantidade balanceadas e Python "**print(_)" para ver como ficaram os dados depois do balanceamento**
   - As amostras para os dados de treino e teste, ficaram com a quantidade de 369 para cada variável de treino e teste

```python
# Aplicação da técnica "under-sampling", rus.fit_resample*

rus = RandomUnderSampler()
rus_X, rus_y = rus.fit = rus.fit_resample( train_X, train_y )

print()
print( 'Python imprima relatório "RandomUnderSampler" Balanceamento dos Dados\n' )
print( pd.Series( rus_y ).value_counts() )
print(_)
print()
fig, ax = plt.subplots(figsize=(10,5 ), sharex=True)
sns.countplot( rus_y );
plt.xticks( [ 0, 1 ], [ 'NORMAL', 'FRAUDE' ], fontsize=12, color='#007F5F' );
plt.title( 'VARIÁVEIS BALANCEADAS', pad=15, fontsize=13, color='#007F5F' ) 

plt.tight_layout()
```

### Aplicação do `Modelo` para Treinamento
  - Vou instânciar o mesmo modelo usado, LogisticRegression e farei o (split), separação dos dados para teste em outro modelo
  
  - Utilizarei o modelo treinado, em outro conjunto de dados. Mas, para que haja imparcialidade dos resultados, coerência, preciso fazer com que os novos dados estejam nas mesmas condições e moldes do processo anteior, modelo já treinado.

  - Aplicarei padronização novamente para as variáveis "Time e Amount", mas dessa fez, não farei balanceamento dos dados, porque averá supressão de valores, ou ainda, perdas que cuminarão no enviezamento do modelo. 
  - Aplicarei tratamento, para que ele faça previsões em todas as linhas do conjunto de dados.
  - Feito as parâmetrizações e tratamento para o novo modelo, novamente separo os dados, exluindo as variáveis Time e Amount, porque quem assumirá seus papeis serão as variáveis já balanceadas, std_amount e std_time.
  
  - Imprimirei 3 linhas para vermos com ficam os dados depois das implementações
  
  - Grafico com uma seta indicando o padronizado, eixo ( y=0 )


```python
# Descrição do Modelo para treinamento

test_dfp = data.sample( frac=0.15, random_state=0 )
data = data.drop( test_dfp.index )
```

```python
# Padronização das variáveis "Time e Amount", para Teste

dfp_split = test_dfp.copy()

std_scaler = StandardScaler()
dfp_split[ 'std_amount' ] = std_scaler.fit_transform( dfp_split[ 'Amount' ].values.reshape( -1, 1 ))
dfp_split[ 'std_time' ]   = std_scaler.fit_transform( dfp_split[ 'Time' ].values.reshape( -1, 1 ))

dfp_split.drop([ 'Time', 'Amount' ], axis=1, inplace=True )

dfp_split.iloc[  :3, 5: ]
```

```python
# Plot de dispersão para visualização dos dados ( originais e dados padronizados )

std_scaler = StandardScaler()
std = StandardScaler().fit_transform( data[[ 'Time', 'Amount' ]])

fig, ax = plt.subplots( figsize=(15, 6 )) 
ax.scatter( 'Time', 'Amount', color='#FA003F', s=7, label='Original', alpha=.5, data=data )
ax.scatter( std[:,0], std[:,1], color='#000022', s=7, label='Padronizado', alpha=.5 )

plt.annotate( 'Padronizado', fontsize=7, color='#212529',
              horizontalalignment = 'left',
              verticalalignment = 'bottom',
              xytext = ( -0.0, 7. ),
              xy = (0.7, -7.-1),
              arrowprops = dict(facecolor = '#90E0EF', shrink = 0.20))

plt.legend( loc='upper left' )

plt.title( 'DISPERSÃO DOS DADOS ORIGINAIS E PADRONIZADOS\n', pad=25, fontsize=13, color='#38302E' )

plt.tight_layout()
```
