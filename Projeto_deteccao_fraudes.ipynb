### Criei uma função para descrever melhor as variáveis "sumarydata"**

   - Checagem de dados `ausentes`, dados nulos ( nan ), dados NaN (Not a Number) representam dados missing, que são registros que não existem ou que se perderam dentro do conjunto da base de dados
   
   - Check das variáveis, types (tipos) distribuição dos valores, e nome das variáveis

```python
# Função para otimizar a leitura das entredas e chegar as variáveis

print( 'Python imprima descritivo das variáveis\n' )

def sumarydata ( data ): 
   
    summary = pd.DataFrame( data.dtypes, columns=[ 'dtypes' ] )
    summary = summary.reset_index()
    summary[ 'Nome Variável' ] = summary['index']
    summary = summary[[ 'Nome Variável', 'dtypes' ]]
    summary[ 'Valores Únicos' ] = data.nunique().values
    summary[ 'Valores Primários' ]  = data.loc[0].values
    summary[ 'Valores Secudários' ] = data.loc[1].values
    summary[ 'Dados Nulos' ] = data.isnull().sum().values
    summary[ 'Dados Nulos (%)' ] = ((( data.isnull().sum() / data.shape[0] ) * 100 ).sort_values(ascending = False ))
    
    return summary

check = sumarydata ( data )
check.head(5)
```

### Método matemático **PCA** para Python

  - Aplicação de uma técnica para reduzir a dimensão do DataFrame, ( base de dados ), preservando suas propriedades e integridade dos dados.
   - Vejamos os recursos gerados por este método, como cada recurso se correlaciona com a classe desse DataFrame.

  - Origem e Aplicação: PCA, é um modelo de procedimento `matemático` que utiliza a transformação `ortogonal`, (ortogonalização de vetores) para converter um conjunto de observações de variáveis possivelmente correlacionadas, em conjunto de valores de variáveis `linearmente` não correlacionadas chamadas de componentes principais ).
  
  O método PCA nos permite a redução da dimensionalidade enquanto mantém o maior número possível de informações. Para conseguir isso, o algoritmo encontra um conjunto novo de recursos - os chamados **`componentes`**.

Esses componentes são em número menor or igual às variáveis originais. No caso deste projeto, os componentes achados pela transformação da PCA são as próprias colunas **[ 'V28', 'V27', 'V26', 'V25', 'V24', 'V23', 'V22', 'V20', 'V15', 'V13', 'V8' ]**.

  - Os gráficos abaixo, mostram o comportamento e distribuição de cada uma das variáves e sua importância para o processo de predição de fraudes.
   - Quanto mais distintas forem umas das outras, melhor, assim podemos identificar fraudes mais facilmente. Veja como exemplo as variáveis `V3`, `V4`, `V10` e `V11` e `V28`, são variáveis com uma certa distinção de comportamento.

   - Por outro lado, variáveis como `V8`, `V13` e `V15` são muito similares, apresentando aproximadamente o mesmo comportamento, nada bom.

```python
# Aplicação da técnica "under-sampling", rus.fit_resample*

rus = RandomUnderSampler()
rus_X, rus_y = rus.fit = rus.fit_resample( train_X, train_y )

print()
print( 'Python imprima relatório "RandomUnderSampler" Balanceamento dos Dados\n' )
print( pd.Series( rus_y ).value_counts() )
print(_)
print()
fig, ax = plt.subplots(figsize=(10,5 ), sharex=True)
sns.countplot( rus_y );
plt.xticks( [ 0, 1 ], [ 'NORMAL', 'FRAUDE' ], fontsize=12, color='#CC5803' );
plt.title( 'VARIÁVEIS BALANCEADAS', pad=15, fontsize=13, color='#CC5803' ) 

plt.tight_layout()
```
